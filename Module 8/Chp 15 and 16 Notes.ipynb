# Plotting a Simple Graph

```python
import matplotlib.pyplot as plt

squares = [1,4,9,16,25]
plt.plot(squares)
plt.show()
```

# Changing the Label Type and Graph Thickness

```python
squares = [1,4,9,16,25]
plt.plot(squares, linewidth=5)

#Set chart title and label axes.
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

#Set size of tick labels.
plt.tick_params(axis='both',labelsize=14)

plt.show()
```

#  Correcting the Plot (square of 4 is not 25!)

When plotting, if you just give a sequence of numbers, Python will assume the first data point corresponds to x-coordinate of 0. But here, we want the x-coordinate to be 1.

```python
input_values = [1,2,3,4,5]
squares = [1,4,9,16,25]
plt.plot(input_values, squares, linewidth=5)

#Set chart title and label axes.
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

#Set size of tick labels.
plt.tick_params(axis='both',labelsize=14)

plt.show()
```

# Plotting and Styling Individual Points with scatter()

```python
plt.scatter(2,4)
plt.show
```

```python
plt.scatter(2,4,s=200)

#Set chart title and label axes.
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

#Set size of tick labels.
plt.tick_params(axis='both', which='major', labelsize=14)

plt.show()
```

# Plotting a Series of Points with scatter()

```python
x_values = [1,2,3,4,5]
y_values = [1,4,9,16,25]

plt.scatter(x_values,y_values,s=100)

#Set chart title and label axes.
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

#Set size of tick labels.
plt.tick_params(axis='both', which='major', labelsize=14)

plt.show()
```

# Calculating Data Automatically

Writing out lists by hand can be inefficient. Use a loop to do calculations for you. Here's with 1000 points.

```python
plt.rcParams['figure.figsize'] = [20, 10]

x_values = list(range(1,1001))
y_values = [x**2 for x in x_values]

plt.scatter(x_values,y_values,s=40)

#Set chart title and label axes.
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

#Set size of tick labels.
plt.tick_params(axis='both', which='major', labelsize=14)

#Set the range for each axis.
plt.axis([0,1100,0,1100000])

plt.show()
```

# Removing Outlines from Data Points

matplotlib lets you color points individually in a scatter plot. The default is blue dots with a black outline. This works well with just a few points, but blends together with many points. To remove the outlines around the points, use **edgecolor='none'** when you call **scatter()**.

```python
plt.rcParams['figure.figsize'] = [20, 10]

x_values = list(range(1,1001))
y_values = [x**2 for x in x_values]

plt.scatter(x_values,y_values,edgecolor='none',s=40)

#Set chart title and label axes.
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

#Set size of tick labels.
plt.tick_params(axis='both', which='major', labelsize=14)

#Set the range for each axis.
plt.axis([0,1100,0,1100000])

plt.show()
```

# Defining Custom Colors

Define specific colors with **c='red'**, for example. 

Or, you can use the RGB color model, as in **plt.scatter(x_values,y_values,c=(0,0,0.8),edgecolor='none',s=40)**. One decimal value each for three colors - red, green, and blue. Use values between 0 and 1. Values closer to 0 produce dark colors. Values closer to 1 produce lighter colors. This example shows light blue dots. 

```python
plt.rcParams['figure.figsize'] = [20, 10]

x_values = list(range(1,1001))
y_values = [x**2 for x in x_values]

plt.scatter(x_values,y_values,c='red',edgecolor='none',s=40)

#Set chart title and label axes.
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

#Set size of tick labels.
plt.tick_params(axis='both', which='major', labelsize=14)

#Set the range for each axis.
plt.axis([0,1100,0,1100000])

plt.show()
```

# Using a Colormap

A *colormap* is a series of colors in a graduate that moves from a starting to ending color. They're used in visualizations to emphasize a pattern in the data, such as low values being a light color and high values being a darker color. 

```python
plt.rcParams['figure.figsize'] = [20, 10]

x_values = list(range(1,1001))
y_values = [x**2 for x in x_values]

plt.scatter(x_values,y_values,c=y_values,cmap=plt.cm.Blues,edgecolor='none',s=40)

#Set chart title and label axes.
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

#Set size of tick labels.
plt.tick_params(axis='both', which='major', labelsize=14)

#Set the range for each axis.
plt.axis([0,1100,0,1100000])

plt.show()
```

# Saving Your Plots Automatically

```python
plt.rcParams['figure.figsize'] = [20, 10]

x_values = list(range(1,1001))
y_values = [x**2 for x in x_values]

plt.scatter(x_values,y_values,c=y_values,cmap=plt.cm.Blues,edgecolor='none',s=40)

#Set chart title and label axes.
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

#Set size of tick labels.
plt.tick_params(axis='both', which='major', labelsize=14)

#Set the range for each axis.
plt.axis([0,1100,0,1100000])

plt.show()

plt.savefig('squares_plot.png', bbox_inches='tight')
```
